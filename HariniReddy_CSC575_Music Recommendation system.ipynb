## 1. Hyperlink Analysis
## A

```python
# Initialization
alpha = 0.15
pages = ['A', 'B', 'C']
links = {'A': ['B', 'C'], 'B': ['C'], 'C': ['A']} 
R = {page: 1/len(pages) for page in pages} 

def calculate_new_ranks(R, links, alpha):
    new_R = {}
    S = len(R)
    for page in R:
        # Summation part of the formula
        sum_part = sum(R[q] / len(links[q]) for q, links_q in links.items() if page in links_q)
        new_R[page] = alpha / S + (1 - alpha) * sum_part
    return new_R
```

```python

def normalize_ranks(R):
    norm_factor = 1 / sum(R.values())
    return {page: rank * norm_factor for page, rank in R.items()}


iterations = 3
results = []

```

```python
for i in range(iterations):
    new_R = calculate_new_ranks(R, links, alpha)
    normalized_R = normalize_ranks(new_R)
    results.append((new_R, normalized_R))
    R = normalized_R
```

```python
formatted_results = ""

for i, (before_norm, after_norm) in enumerate(results, start=1):
    formatted_results += f"Iteration {i}:\n"
    formatted_results += "Before Normalization:\n"
    for page, score in before_norm.items():
        formatted_results += f"  - Page {page}: {score:.4f}\n"
    formatted_results += "After Normalization:\n"
    for page, score in after_norm.items():
        formatted_results += f"  - Page {page}: {score:.4f}\n"
    formatted_results += "\n"

print(formatted_results)
```

```python
pages1 = ['A', 'B', 'C', 'D']
links1 = {'A': ['B', 'C'], 'B': ['C', 'D'], 'C': ['B'], 'D': ['C']} 
authority = {page: 1 for page in pages1}
hub = {page: 1 for page in pages1}

def update_scores(links, authority, hub):
    new_authority = {}
    new_hub = {}
    #  authority scores
    for p in pages1:
        new_authority[p] = sum(hub[q] for q, links_q in links1.items() if p in links_q)
    #  hub scores
    for p in pages1:
        new_hub[p] = sum(authority[q] for q in links1[p]) if p in links1 else 0
    
    return new_authority, new_hub

def normalize_scores(scores):
    norm = sum(value ** 2 for value in scores.values()) ** 0.5
    return {page: score / norm for page, score in scores.items()}

```

```python
iterations = 3

for i in range(iterations):
    authority, hub = update_scores(links1, authority, hub)
    
    # Print before normalization
    print(f"Iteration {i+1} Before Normalization:")
    print("Authority:", authority)
    print("Hub:", hub)
    
    # Normalize scores
    authority = normalize_scores(authority)
    hub = normalize_scores(hub)
    
    # Print after normalization
    print(f"Iteration {i+1} After Normalization:")
    print("Authority:", authority)
    print("Hub:", hub)
    print("\n")
```

```python
import pandas as pd

spotify_metadata = pd.read_csv('Spotify_Metadata.csv')
spotify_numeric = pd.read_csv('Spotify_Numeric.csv')

with open('Spotify_Users.txt', 'r') as file:
    spotify_users = file.readlines()
    spotify_users = [list(map(int, line.strip().split(','))) for line in spotify_users]

# VIEW 
spotify_metadata.head(), spotify_numeric.head(), spotify_users[:2]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# cosine similarity function
def cosine_sim(vec1, vec2):
    
    vec1 = np.array(vec1).reshape(1, -1)
    vec2 = np.array(vec2).reshape(1, -1)
    return cosine_similarity(vec1, vec2)[0][0]

```

```python
def knn(item, Data, K):
    
    # Compute cosine similarity 
    similarities = cosine_similarity([item], Data)[0]
    
    sorted_indices = np.argsort(similarities)[::-1]
    
    # Select the top K indices and their corresponding similarity values
    top_k_indices = sorted_indices[:K]
    top_k_similarities = similarities[top_k_indices]
    
    # Include metadata for the recommended songs
    recommendations = spotify_metadata.loc[top_k_indices, ['name', 'artist', 'mood(s)']]
    recommendations['similarity'] = top_k_similarities
    
    return recommendations.reset_index()

# Demonstrate the knn function with the specified feature vector and K=10
item_vector = [0.8, 0.1, 0.7, 0.0, 0.5, 0.1, 0.5, 0.2, 0.0, 1.0, 0.1, 0.2, 0.7]
knn_results = knn(item_vector, spotify_numeric.drop('index', axis=1).values, 10)
knn_results
```

```python
def recommend(userID, Data, K):
    
    # get user's playlist
    playlist_indices = spotify_users[userID]
    
    playlist_features = Data.iloc[playlist_indices].drop('index', axis=1).mean().values
    
    # Use knn 
    all_recommendations = knn(playlist_features, Data.drop('index', axis=1).values, K + len(playlist_indices))
    
    # Filtering out songs already in the user's playlist
    recommendations = all_recommendations[~all_recommendations['index'].isin(playlist_indices)].head(K)
    
    return recommendations

```

```python
# Converting Spotify_Numeric.csv to DataFrame for easier manipulation
spotify_numeric_df = pd.DataFrame(spotify_numeric)

# Show results 
recommendations_user_3 = recommend(3, spotify_numeric_df, 20)
recommendations_user_19 = recommend(19, spotify_numeric_df, 20)
print("Recommendations for User 3:")
recommendations_user_3
```
